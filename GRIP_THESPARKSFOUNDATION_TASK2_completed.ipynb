```python
df1=df.drop(['Species'],axis=1)
df1
```

### 
we can see that sepalwidth has some outliers

```python
#finding optimal number of clusters using elbow method
from sklearn.cluster import KMeans
#sum of squared distance
ssd=[]
range_n=[2,3,4,5,6,7,8,9]
for i in range_n:
    kmeans = KMeans(n_clusters=i,max_iter=150,random_state=42)
    kmeans.fit(df1)
    ssd.append([i,kmeans.inertia_])
ssd_dataframe=pd.DataFrame(ssd)
plt.plot(ssd_dataframe[0],ssd_dataframe[1],'go--', color='purple')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Within clusters sum of square')
plt.grid()
plt.show()
```

```python
#creating K-Means classifier model 
#assuming k=3

Kmeans=KMeans(n_clusters=3,max_iter=500,random_state=50)
y_kmeans=Kmeans.fit_predict(df1)
y_kmeans
```

```python
# Visualizing clusters using PetalLengthCm and PetalWidthCm columns

plt.scatter(df.loc[y_kmeans == 0, 'PetalLengthCm'], df.loc[y_kmeans == 0, 'PetalWidthCm'], s=100, marker='*', c='blue', label='Iris-versicolor')
plt.scatter(df.loc[y_kmeans == 1, 'PetalLengthCm'], df.loc[y_kmeans == 1, 'PetalWidthCm'], s=100, marker='*', c='red', label='Iris-setosa')
plt.scatter(df.loc[y_kmeans == 2, 'PetalLengthCm'], df.loc[y_kmeans == 2, 'PetalWidthCm'], s=100, marker='*', c='green', label='Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], s=300, marker='*', c='yellow', label='Centroids')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Clustering - Iris Dataset')
plt.legend()
plt.show()
```
